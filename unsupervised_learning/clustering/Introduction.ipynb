One one axis, you have flat vs heirarchical clustering algorithms; on the other centroid/parametric vs density based/non-parametric

1) Flat centroid based - This includes algorithms like K-Means. This is O(N), but very sensitive to the choice of initial conditions

Works for clean data , small amounts, low dimensions, easy to visualize 


But not for more complex shapes  


They don't work well at all, as K-means make the gaussian ball assumption (clusters are high dimensional spheres)


Here are other examples where K-means does not work




Where you have two concentric groups of points, but K-means fails to identify this


or this



Where if clusters have uneven number of points (one cluster has very few point , for example), K-means gives weights to the clusters with more points, and is perfectly happy placing clusters with very few points far away



2. Flat , Density based methods - DBSCAN (density based spatial clustering for applications with noise) - refer the accompanying notebook. Note that the complexity of DBSCAN is O(n^2) in worse case, as for every point, it checks an episolon neighbourhood of the data

3. Flat, heirarchical methods : Example : ward , complete linkage

What is commonly called heirarchical clustering, or heirarchical agglomerative clustering (HAC)

[This](https://stats.stackexchange.com/questions/195446/choosing-the-right-linkage-method-for-hierarchical-clustering) article has information of differnet kinds of heirarchical clustering approaches


The basic method of HAC algorithm is  generic; it amounts to updating, at each step, by the formula known as Lance-Williams formula, the proximities between the emergent (merged of two) cluster and all the other clusters (including singleton objects) existing so far. There exist implementations not using Lance-Williams formula. But using it is convenient: it lets one code various linkage methods by the same template.

Different approaches differ in how proximity between any two clusters is defined at every step

Some examples, there are a lot more, for more details, check linkage above

i) Single linkage - Use nearest neighbor
ii) Complete linkage - Use farthest neighbour
iii) Ward's method - Proximity between two clusters is determined as the magnitude by which summed square of then joint cluster is greater than sum of squares of each of the two clusters taken separately  (ie S1S2 - (S1 + S2))



There are also caveats about how to interpret HAC

https://stats.stackexchange.com/questions/63546/comparing-hierarchical-clustering-dendrograms-obtained-by-different-distances/63549#63549

Some are

i) Never compare visually clusters obtained by different distance methods, as they make different assumptions
ii) In ward's method, you cannot "cut" at a level and determine the number of clusters - This doesn't work because ward uses summation and not average, meaning that later clusters which have more points look misleadingly better, etc
