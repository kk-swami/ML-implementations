# Grouped Query Attention

Used in LlAMA-2 , LlAMA-1 uses regular attention




https://deci.ai/blog/decilm-15-times-faster-than-llama2-nas-generated-llm-with-variable-gqa/

Let's think about the dimensions a little bit

In regular ,multi-headed attention,

Let's say X is matrix of dimension T * d  ## T Tokens, each of d dimensions, for one sample sentence X

Let's say we have H heads, each of dimension h

So taking one head, for one sample X,

Q = X * Wq to give dimension T * h (Wq is a weight matrix of d * h dimensions)
K = X * Wk to give dimension T * h (Wk is a weight matrix of d * h dimensions)

The weight matrix = softmax(Q * transpose(K)/ sqrt(h)) = T * T dimension

V = X * Wv to give dimension T * h (Wv is a weight matrix of d * h dimensions)

We concatenate H  T * h matrices to get T * d (same dimension as X)( which goes to feed forward layer



So thinking about the parameters, for every head, we have Wq, Wk and Wv matrices



Im multi query attention, you use full parameter sharing across the H heads for the key and value matrices

So you still have H Query matrices ( H Wq matrices), but only 1 key and 1 value matrix

Im grouped query attention, you use parameter sharing for some heads for the key and value matrices, not all

So you still have H Query matrices ( H Wq matrices), but m Key and m Query matrices, where m < H
