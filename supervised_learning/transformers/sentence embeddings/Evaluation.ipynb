# Evaluation of search 

This evaluation can refer to embeddings trained using sentence transformers or OpenAI Language models or others;
or even plain old lexical search BM25

## The need for benchmarks

You can't improve what you can't measure

Nils Reimer strongly argues that many benchmarks cherrypick, and it is easy to overfit to a benchmark, hence the need to keep coming up and evaluating on newer more complex benchmarks



## STS - Sentence text similarity

Nils again argues that this metric doesn't have any predictability

STS - Given 2 sentences, a human has to annotate a label between 0 and 5. 5 - High similarity, 0, low similarity



## BEIR (Benchmarking IR) - benchmark to evaluate embeddings on out of domain also, not only in domain

Traditionally, embeddings are evaluated on the same domain they are trained on (maybe ok for most real life applications ?? )

These guys say that for generalization, embeddings trained on one domain should be tested on a different domain ?? 

And then we find BM25 is a great baseline. A lot of models (DPR - dense passage retrieval from facebook, performs worse than the baseline of BM25 for 17 datasets tested, and performs better only for 1 dataset)

BM25 for retrieval followed for using Cross encoder for reranking worked the best !!

Note though that this was in 2021 ! BM25 was better than dense models such as DPR, TAS-B for 10/18 datasets
In 2023, BM25 is better only for 2/18 datasets. Need for new benchmarking ?

## MTEB (Massive Text Embedding Benchmark)

2022 - 
MTEB spans 8 embedding tasks (clustering, bitext mining, retrieval, STS, classification, summarization, reranking, pair classification) covering a total of 58 datasets and 112 languages



Evaluated 33 models 

MTEB score on y axis, speed in examples/sec on X axis (higher speed better)
The size of the circle is indicative of the size of the embedding

On left side, 6B parameter GPT models, which have good MTEB score, but very low speed, can you use something which is faster such as MPNet which doesn't reduce accuracy too much ?lu
